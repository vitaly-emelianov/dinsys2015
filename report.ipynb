###Описание

Построить визуальную модель, отражающую процесс кодирования и диаграммы кодов для преобразований перекладывания 2 и 3 отрезков (в случае двух отрезков система изоморфна повороту окружности). Необходимо особо учесть случай рациональных длин отрезков.

###Программный код

* кодирование преобразования перекладывания отрезков реализовано в классе View, который находится в файле view.py
* класс View реализует метод processStep(), который вычисляет один шаг преобразования и  список смежности дерева кодов (атрибут tree) на этом шаге
* визуализация дерева осуществляется в функции visualize(adjacency_list) файла visualization.py при помощи сторонних библиотек языка Python: 
       matplotlib 
       networkx 
       pygraphviz
Уточним, что перекладывание реализовано как циклический сдвиг вправо на длину самого правого отрезка.

###Демонстрация работы 

```python
from view import View
from visualization import visualize
import math
%matplotlib inline
```

####Рациональные длины отрезков
Посмотрим на дерево кодов, когда длины отрезков равны соответственно $\frac{1}{3}, \frac{1}{2}, \frac{1}{6}$

```python
init_codes = {
    (0., 1./3): 'a',
    (1./3, 5./6): 'b',
    (5./6, 1.): 'c'
}
view = View(init_codes, 1./6);

for x in xrange(1, 10):
    view.processStep()
visualize(view.tree, labels=True)
```

```python
for x in xrange(1, 80):
    view.processStep()
visualize(view.tree, custom_alpha=1)
```

####Иррациональные длины отрезков
Посмотрим на дерево кодов, когда длина самого правого отрезка равна $\frac{3 + \sqrt 5}{2}$

```python
alpha = (3. - math.sqrt(5.)) / 2.
init_codes = {
    (0., 1. - 2. * alpha): 'a',
    (1. - 2. * alpha, 1. - alpha): 'b',
    (1. - alpha, 1.): 'c'
}
view = View(init_codes, alpha);

for x in xrange(1, 10):
    view.processStep()
visualize(view.tree, labels=True)
```

```python
for x in xrange(1, 50):
    view.processStep()
visualize(view.tree, custom_alpha=1)
```

```python
for x in xrange(1, 200):
    view.processStep()
visualize(view.tree, custom_alpha=1)
```
